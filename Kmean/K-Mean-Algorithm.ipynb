# K-mean using Python Libraries

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('8-Kmeans_EM.csv')
```

```python
data.head()
```

```python
f1=data['V1'].values
f2=data['V2'].values
x=np.array(list(zip(f1,f2)))
```

```python
plt.scatter(f1,f2,color='black',s=10,label="original data")
plt.legend()
plt.xlabel('X')
plt.ylabel("Y")
```

```python
k=3
kmean=KMeans(n_clusters=k)
kmean=kmean.fit(x)
label=kmean.predict(x)
centroid=kmean.cluster_centers_
print(centroid)
plt.scatter(x[:,0],x[:,1],c=label,s=5,cmap='rainbow',label="orignial data")
plt.scatter(centroid[:,0],centroid[:,1],color='black',marker='*',s=60,label="centroid")
plt.legend()
```

# Using k-mean algorithms

```python
# random centroid
center=np.array([[f1[0],f2[0]],[f1[1],f2[1]],[f1[2],f2[2]]])
```

```python
# inital centroid
center
```

```python
# data for cluster
x
```

```python
from sklearn.metrics import pairwise_distances_argmin
```

```python
# running Algorithms
while True:
    labels=pairwise_distances_argmin(x,center)
    new_centers=np.array([x[labels==i].mean(0) for i in range(3)])
    if np.all(center==new_centers):
        break
    center=new_centers
```

```python
center
```

```python
plt.scatter(f1,f2,c=label,s=5,cmap='rainbow')
plt.scatter(center[:,0],center[:,1],c='black')
```
