```python
from sklearn.datasets import fetch_openml
import numpy as np
```

```python
train_data=data[0:5000]
```

```python
train_target=target[0:5000]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,train_target,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(random_state=0)
# fitting to model
logistic=logistic.fit(x_train,y_train)
```

```python
# prediction of the data on behalf of the Regression
predict=logistic.predict(x_test)
```

```python
for i in range(len(x_test)):
    print(y_test[i]," ",predict[i])
```

# predicit accuract

```python
from sklearn import metrics
print("accuracy :",metrics.accuracy_score(y_test,predict))
```

```python
#convert the tagert  to 0 or 1
#if target==5 make 1 else 0
target_data=(mnist['target']=='5').astype(np.int)
```

```python
target_data=target_data[0:5000]
```

```python
train_data
```

```python
target_data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,target_data,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(random_state=0)
# fitting to model
logistic=logistic.fit(x_train,y_train)
```

```python
y_predict=logistic.predict(x_test)
```

```python
y_predict
```

```python
from sklearn import metrics
print("accuracy :",metrics.accuracy_score(y_test,y_predict))
```
